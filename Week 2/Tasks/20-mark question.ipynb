```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Pandas CSV Handling & Visualization

```python
dataset = pd.read_csv('salaries.csv')
```

```python
dataset.head(5)
```

```python
# Check for missing data
dataset.isna().sum()

# There are no missing data in the dataset
```

```python
average_salary_group = dataset.groupby('job_title')['salary_in_usd'].mean()
```

```python
plt.figure(figsize=(54, 10), facecolor='#eee')

plt.bar(average_salary_group.index, average_salary_group.values, color='coral', width=0.5)
plt.xticks(rotation=45, ha='right')

plt.show()
```
