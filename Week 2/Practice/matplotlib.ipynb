```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1, 8 , 27, 64, 125]

plt.xlabel('x')
plt.ylabel('y')
plt.title('y = x^3')
plt.grid()

plt.plot(x, y, color='green', linestyle='dashed', linewidth = 1, marker='o', markerfacecolor='green', markersize=5)

plt.show()
```

```python
x = [15,55,78]

plt.pie(x, labels=['A', 'B', 'C'], colors=['orange', 'green', 'pink'])
plt.show()
```

```python
problems = [400, 350, 200]

reasons = ['Bad Code', 'Bad Design', 'Bad Management']

plt.pie(problems, labels=reasons, autopct='%1.1f%%')
plt.show()
```

```python
x = ['A', 'B', 'C']
y = [10, 15, 30]

plt.bar(x, y, color = ['magenta', 'orange', 'green'])
plt.show()
```
